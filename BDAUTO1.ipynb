**Особенности работы с категориальными показателями в составе
таблицы данных**

```python
import pandas as pd
shop = {"gender":pd.Series(["female", "male", "male", "female", "male", "female"]),
 "size":pd.Series(["XS", "L", "M", "XL", "XS", "S"]),
 "totalsum":pd.Series([3000, 2456, 8965, 4567, 3487, 955])}
shop_df = pd.DataFrame(shop)
shop_df
```

```python
#Преобразуем столбец показателя пол (gender) в числовые метки
shop_df["gender"]= pd.Categorical(shop_df["gender"])
shop_df["gender_cat"]=shop_df["gender"].cat.codes
shop_df
```

```python
#Преобразуем столбец показателя размер (size) в числовые метки
shop_df["size"]= pd.Categorical(shop_df["size"], categories = ["XS", "S", "M", "L", "XL"])
shop_df["size_cat"]=shop_df["size"].cat.codes
shop_df
```

`В базе данных автомобилей присутствуют такие данные как расход гсм, мощность (в л.с.), ускорение, - представленные в Float64 — это встроенный тип данных в Python, который представляет числа с плавающей запятой с двойной точностью, вес и количество цилиндров - int64 - целочисленные значения, а также наименование и страна-производитель - класс object`

```python
#Извлечение данных из набора automobile.csv всех записей, в которых страна-производитель - Япония и мощность боллее 100 л.с.
#ОТБОР ДАННЫХ ПО ЛОГИЧЕСКОМУ УСЛОВИЮ
Automobile_df[(Automobile_df['origin']=="japan") & (Automobile_df["horsepower"]>=100)]
```

```python
Japan_df=Automobile_df[(Automobile_df['origin']=="japan") & (Automobile_df["horsepower"]>=100)]
```

```python
Japan_df[["name","model_year"]]
```

```python
#По японскому списку - мощность - по убыванию, ускорение - по возрастанию. Топ 6
Japan_ord=Japan_df.sort_values(by=["horsepower", "acceleration"], ascending=[False,True])
Japan_df.sort_values(by=["horsepower", "acceleration"], ascending=[False,True]).head(6)
```

```python
#Проверка изменчивости исходного списка
Japan_df.head(6)
```

```python
Japan_ord.reset_index(inplace=True, drop=True)
Japan_ord
Japan_ord.head(3)
#INPLACE - заменяет первоначальные индексы
#DROP - удаляет данные о первоначальном индексе
```
